```python
# Install libraries
!pip install scikit-image opencv-python pillow gradio imbalanced-learn seaborn --quiet

# Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import os
import tempfile
from skimage import feature, measure
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
from imblearn.over_sampling import SMOTE
import joblib
from PIL import Image
import gradio as gr
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ Libraries loaded")

# =============================================================================
# SETUP
# =============================================================================
def setup_data():
    from google.colab import drive
    drive.mount('/content/drive')

    base_path = "/content/drive/My Drive/breast c"
    us_path = os.path.join(base_path, "Dataset_BUSI_with_GT")
    mg_path = os.path.join(base_path, "MIASDBv1.21")

    if not os.path.exists("ultrasound_dataset") and os.path.exists(us_path):
        os.symlink(us_path, "ultrasound_dataset")
        print("‚úÖ Ultrasound dataset linked")

    if not os.path.exists("mammography_dataset") and os.path.exists(mg_path):
        os.symlink(mg_path, "mammography_dataset")
        print("‚úÖ Mammography dataset linked")

setup_data()

# Check datasets
def check_data(path, name):
    classes = ['normal', 'benign', 'malignant']
    counts = {}

    for cls in classes:
        cls_path = os.path.join(path, cls)
        if os.path.exists(cls_path):
            images = [f for f in os.listdir(cls_path)
                     if f.lower().endswith(('.png', '.jpg', '.jpeg', '.pgm'))]
            counts[cls] = len(images)
        else:
            counts[cls] = 0

    total = sum(counts.values())
    print(f"üìä {name}: {total} images")
    for cls, count in counts.items():
        print(f"   {cls}: {count} ({count/total*100:.1f}%)")

    return total

print("\nüìä Dataset Summary:")
us_total = check_data("ultrasound_dataset", "Ultrasound")
mg_total = check_data("mammography_dataset", "Mammography")

# =============================================================================
# FEATURE EXTRACTION
# =============================================================================
def extract_features(image_path, modality):
    try:
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            return None

        size = (256, 256) if modality == "ultrasound" else (512, 512)
        img = cv2.resize(img, size)

        img = cv2.medianBlur(img, 5)
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
        img = clahe.apply(img)

        features = {}

        features['mean'] = np.mean(img)
        features['std'] = np.std(img)
        features['entropy'] = measure.shannon_entropy(img)

        glcm = feature.graycomatrix(img, [1, 3], [0, np.pi/4, np.pi/2], symmetric=True, normed=True)

        features['contrast'] = np.mean(feature.graycoprops(glcm, 'contrast'))
        features['homogeneity'] = np.mean(feature.graycoprops(glcm, 'homogeneity'))
        features['energy'] = np.mean(feature.graycoprops(glcm, 'energy'))
        features['correlation'] = np.mean(feature.graycoprops(glcm, 'correlation'))

        _, thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        if contours:
            largest = max(contours, key=cv2.contourArea)
            area = cv2.contourArea(largest)
            perimeter = cv2.arcLength(largest, True)
            features['circularity'] = (4 * np.pi * area) / (perimeter ** 2) if perimeter > 0 else 0
        else:
            features['circularity'] = 0

        edges = cv2.Canny(img, 50, 150)
        features['edge_density'] = np.sum(edges > 0) / edges.size

        return features

    except Exception as e:
        print(f"Error: {e}")
        return None

# =============================================================================
# TRAINING
# =============================================================================
def train_model(dataset_path, modality_name, save_name):
    print(f"\nüéØ Training {modality_name} model...")

    dataset = []
    classes = ['normal', 'benign', 'malignant']

    for class_name in classes:
        class_path = os.path.join(dataset_path, class_name)
        if not os.path.exists(class_path):
            continue

        images = [f for f in os.listdir(class_path)
                 if f.lower().endswith(('.png', '.jpg', '.jpeg', '.pgm'))]

        for img_file in images:
            img_path = os.path.join(class_path, img_file)
            features = extract_features(img_path, modality_name.lower())

            if features is not None:
                features['class'] = class_name
                dataset.append(features)

    if len(dataset) < 30:
        print(f"‚ùå Not enough data: {len(dataset)} samples")
        return None, None, None, None, None

    df = pd.DataFrame(dataset)
    print(f"‚úÖ Total samples: {len(df)}")

    feature_cols = [col for col in df.columns if col != 'class']
    X = df[feature_cols]
    y = df['class']

    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    # 80/20 split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
    )

    print(f"\nüìä Split:")
    print(f"  Train: {X_train.shape[0]} samples")
    print(f"  Test:  {X_test.shape[0]} samples")

    # Balance training data
    smote = SMOTE(random_state=42)
    X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

    # Scale
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_balanced)
    X_test_scaled = scaler.transform(X_test)

    # Model
    model = RandomForestClassifier(
        n_estimators=200,
        max_depth=20,
        min_samples_split=3,
        min_samples_leaf=1,
        max_features='sqrt',
        class_weight='balanced',
        random_state=42,
        n_jobs=-1
    )

    # Cross-validation
    cv_scores = cross_val_score(model, X_train_scaled, y_train_balanced, cv=5)
    print(f"üìä CV Accuracy: {cv_scores.mean():.3f} ¬± {cv_scores.std():.3f}")

    # Train
    model.fit(X_train_scaled, y_train_balanced)

    # Evaluate
    y_pred = model.predict(X_test_scaled)
    y_probs = model.predict_proba(X_test_scaled)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')

    print(f"\nüìà Test Results:")
    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  F1-Score: {f1:.4f}")
    print(f"\nüìã Classification Report:")
    print(classification_report(y_test, y_pred, target_names=le.classes_, digits=3))

    # Confusion Matrix
    plt.figure(figsize=(6, 5))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title(f'Confusion Matrix - {modality_name}')
    plt.tight_layout()
    plt.show()

    # Feature Importance
    importances = model.feature_importances_
    indices = np.argsort(importances)[::-1]

    plt.figure(figsize=(8, 4))
    plt.title("Feature Importance")
    plt.bar(range(len(feature_cols)), importances[indices])
    plt.xticks(range(len(feature_cols)), [feature_cols[i] for i in indices], rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

    # Save
    joblib.dump(model, f'{save_name}.pkl')
    joblib.dump(scaler, f'{save_name}_scaler.pkl')
    joblib.dump(le, f'{save_name}_encoder.pkl')
    joblib.dump(feature_cols, f'{save_name}_features.pkl')

    print(f"‚úÖ Model saved: {save_name}.pkl")

    return model, scaler, le, feature_cols, {
        'accuracy': accuracy,
        'f1_score': f1,
        'train_samples': X_train_balanced.shape[0],
        'test_samples': X_test.shape[0]
    }

# =============================================================================
# PREDICTION
# =============================================================================
def predict(image, modality):
    try:
        if image is None:
            return "‚ùå Upload an image", None

        if modality == "ultrasound":
            if 'model_us' not in globals() or model_us is None:
                return "‚ùå Ultrasound model not available", None
            model, scaler, encoder, features = model_us, scaler_us, encoder_us, features_us
        else:
            if 'model_mg' not in globals() or model_mg is None:
                return "‚ùå Mammography model not available", None
            model, scaler, encoder, features = model_mg, scaler_mg, encoder_mg, features_mg

        # Process image
        if isinstance(image, str):
            img_array = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
        else:
            img_array = np.array(Image.fromarray(image).convert('L'))

        temp_path = os.path.join(tempfile.gettempdir(), "temp_img.png")
        cv2.imwrite(temp_path, img_array)

        feature_dict = extract_features(temp_path, modality)
        os.remove(temp_path)

        if feature_dict is None:
            return "‚ùå Failed to extract features", None

        # Predict
        feature_vector = np.array([[feature_dict.get(col, 0) for col in features]])
        feature_vector_scaled = scaler.transform(feature_vector)

        prediction_encoded = model.predict(feature_vector_scaled)[0]
        probabilities = model.predict_proba(feature_vector_scaled)[0]
        prediction = encoder.inverse_transform([prediction_encoded])[0]
        confidence = probabilities[prediction_encoded]

        # Result
        colors = {'malignant': '#FF6B6B', 'benign': '#4ECDC4', 'normal': '#45B7D1'}
        color = colors.get(prediction, '#45B7D1')

        result_html = f"""
        <div style="font-family: Arial; padding: 15px;">
            <div style="background: {color}; padding: 15px; border-radius: 8px; color: white; text-align: center;">
                <h2 style="margin: 0;">Prediction: {prediction.upper()}</h2>
                <p style="font-size: 18px;">Confidence: {confidence:.1%}</p>
            </div>

            <div style="margin-top: 15px;">
                <h4>Probabilities:</h4>
        """

        for i, cls in enumerate(encoder.classes_):
            prob = probabilities[i]
            bar_color = colors.get(cls, '#45B7D1')
            result_html += f"""
                <div style="margin: 8px 0;">
                    <div style="display: flex; justify-content: space-between;">
                        <span>{cls.title()}</span>
                        <span>{prob:.1%}</span>
                    </div>
                    <div style="background: #e0e0e0; height: 15px; border-radius: 8px;">
                        <div style="background: {bar_color}; height: 100%; width: {prob*100}%; border-radius: 8px;"></div>
                    </div>
                </div>
            """

        result_html += "</div></div>"

        return result_html, confidence

    except Exception as e:
        return f"‚ùå Error: {str(e)}", None

# =============================================================================
# INTERFACE
# =============================================================================
def create_app():
    with gr.Blocks(title="Breast Cancer Detection", theme=gr.themes.Soft()) as app:

        gr.Markdown("# ü©∫ Breast Cancer Detection")
        gr.Markdown("Upload an ultrasound or mammography image for analysis")

        with gr.Row():
            with gr.Column(scale=1):
                modality = gr.Radio(
                    choices=["ultrasound", "mammography"],
                    label="Image Type",
                    value="ultrasound"
                )
                image_input = gr.Image(
                    label="Upload Image",
                    type="numpy",
                    image_mode="L",
                    height=200
                )
                analyze_btn = gr.Button("üîç Analyze", variant="primary")

            with gr.Column(scale=2):
                output_html = gr.HTML(label="Results")

        analyze_btn.click(
            predict,
            inputs=[image_input, modality],
            outputs=[output_html]
        )

        gr.Markdown("---")
        gr.Markdown("**Note:** This tool is for educational purposes only.")

    return app

# =============================================================================
# MAIN
# =============================================================================
if __name__ == "__main__":
    print("üöÄ Breast Cancer Detection System")
    print("="*40)

    model_us = scaler_us = encoder_us = features_us = None
    model_mg = scaler_mg = encoder_mg = features_mg = None

    # Train ultrasound
    if us_total >= 100:
        print("\n" + "="*40)
        print("Ultrasound Model")
        print("="*40)
        model_us, scaler_us, encoder_us, features_us, _ = train_model(
            "ultrasound_dataset", "Ultrasound", "ultrasound_model"
        )
    else:
        print("‚ö†Ô∏è Not enough ultrasound data")

    # Train mammography
    if mg_total >= 100:
        print("\n" + "="*40)
        print("Mammography Model")
        print("="*40)
        model_mg, scaler_mg, encoder_mg, features_mg, _ = train_model(
            "mammography_dataset", "Mammography", "mammography_model"
        )
    else:
        print("‚ö†Ô∏è Not enough mammography data")

    # Launch
    if model_us is not None or model_mg is not None:
        print("\n‚úÖ Models trained successfully")
        print("\nüöÄ Launching app...")
        app = create_app()
        app.launch(share=True)
    else:
        print("\n‚ùå Failed to train models")
```
