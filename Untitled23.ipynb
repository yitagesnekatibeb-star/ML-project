```python
# =============================================================================
# BREAST CANCER DETECTION - FIXED 80/20 SPLIT VERSION
# =============================================================================

# Install required libraries
!pip install scikit-image opencv-python pillow gradio plotly imbalanced-learn seaborn --quiet

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import os
import tempfile
from skimage import feature, measure
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score, roc_curve
)
from imblearn.over_sampling import SMOTE
import joblib
from PIL import Image
import gradio as gr
import warnings
warnings.filterwarnings('ignore')

print("‚úÖ Libraries imported")

# =============================================================================
# DATASET SETUP - SIMPLIFIED
# =============================================================================
def setup_datasets():
    """Setup datasets from Google Drive - Simplified version"""
    from google.colab import drive
    drive.mount('/content/drive')

    # Define paths
    drive_path = "/content/drive/My Drive/breast c"
    busi_source = os.path.join(drive_path, "Dataset_BUSI_with_GT")
    mias_source = os.path.join(drive_path, "MIASDBv1.21")

    # Create symbolic links instead of copying (faster)
    if not os.path.exists("ultrasound_dataset") and os.path.exists(busi_source):
        os.symlink(busi_source, "ultrasound_dataset")
        print(f"‚úÖ Ultrasound dataset linked")

    if not os.path.exists("mammography_dataset") and os.path.exists(mias_source):
        os.symlink(mias_source, "mammography_dataset")
        print(f"‚úÖ Mammography dataset linked")

# Setup datasets
setup_datasets()

# =============================================================================
# DATA ANALYSIS
# =============================================================================
def analyze_dataset(path, name):
    """Analyze dataset distribution"""
    classes = ['normal', 'benign', 'malignant']
    counts = {}

    for cls in classes:
        cls_path = os.path.join(path, cls)
        if os.path.exists(cls_path):
            images = [f for f in os.listdir(cls_path)
                     if f.lower().endswith(('.png', '.jpg', '.jpeg', '.pgm'))]
            counts[cls] = len(images)
        else:
            counts[cls] = 0

    total = sum(counts.values())
    print(f"üìä {name}: {total} total images")
    for cls, count in counts.items():
        print(f"   {cls.upper()}: {count} ({count/total*100:.1f}%)")

    return counts, total

# Analyze datasets
print("\nüìä DATASET ANALYSIS:")
us_counts, us_total = analyze_dataset("ultrasound_dataset", "Ultrasound")
mg_counts, mg_total = analyze_dataset("mammography_dataset", "Mammography")

# =============================================================================
# ENHANCED FEATURE EXTRACTION
# =============================================================================
def extract_enhanced_features(image_path, modality):
    """Extract comprehensive features from medical images"""
    try:
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            return None

        # Resize for consistency
        size = (256, 256) if modality == "ultrasound" else (512, 512)
        img = cv2.resize(img, size)

        # Enhanced preprocessing
        img = cv2.medianBlur(img, 5)
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
        img = clahe.apply(img)

        features = {}

        # Intensity features
        features['mean_intensity'] = np.mean(img)
        features['std_intensity'] = np.std(img)
        features['entropy'] = measure.shannon_entropy(img)

        # Texture features (GLCM)
        distances = [1, 3]
        angles = [0, np.pi/4, np.pi/2]
        glcm = feature.graycomatrix(img, distances, angles, symmetric=True, normed=True)

        features['contrast'] = np.mean(feature.graycoprops(glcm, 'contrast'))
        features['homogeneity'] = np.mean(feature.graycoprops(glcm, 'homogeneity'))
        features['energy'] = np.mean(feature.graycoprops(glcm, 'energy'))
        features['correlation'] = np.mean(feature.graycoprops(glcm, 'correlation'))

        # Shape features
        _, thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        if contours:
            largest = max(contours, key=cv2.contourArea)
            area = cv2.contourArea(largest)
            perimeter = cv2.arcLength(largest, True)
            features['circularity'] = (4 * np.pi * area) / (perimeter ** 2) if perimeter > 0 else 0
        else:
            features['circularity'] = 0

        # Edge features
        edges = cv2.Canny(img, 50, 150)
        features['edge_density'] = np.sum(edges > 0) / edges.size

        return features

    except Exception as e:
        print(f"Feature extraction error: {e}")
        return None

# =============================================================================
# VISUALIZATION FUNCTIONS
# =============================================================================
def plot_confusion_matrix(y_true, y_pred, classes, title="Confusion Matrix"):
    """Plot confusion matrix with heatmap"""
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes)
    plt.title(title, fontsize=14)
    plt.ylabel('True Label', fontsize=12)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.tight_layout()
    plt.show()

def plot_roc_curves(y_true, y_probs, classes, title="ROC Curves"):
    """Plot ROC curves for multiclass classification"""
    n_classes = len(classes)
    fig, ax = plt.subplots(figsize=(8, 6))

    colors = ['blue', 'green', 'red']

    for i, color in zip(range(n_classes), colors):
        y_true_binary = (y_true == i).astype(int)
        y_prob_binary = y_probs[:, i]

        fpr, tpr, _ = roc_curve(y_true_binary, y_prob_binary)
        auc_score = roc_auc_score(y_true_binary, y_prob_binary)

        ax.plot(fpr, tpr, color=color, lw=2,
                label=f'{classes[i]} (AUC = {auc_score:.3f})')

    ax.plot([0, 1], [0, 1], 'k--', lw=2, label='Random (AUC = 0.5)')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate', fontsize=12)
    ax.set_ylabel('True Positive Rate', fontsize=12)
    ax.set_title(title, fontsize=14)
    ax.legend(loc="lower right")
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_feature_importance(model, feature_names):
    """Plot feature importance"""
    importances = model.feature_importances_

    # Sort features by importance
    indices = np.argsort(importances)[::-1]

    plt.figure(figsize=(10, 6))
    plt.title("Feature Importances", fontsize=14)
    bars = plt.bar(range(len(feature_names)), importances[indices], align="center")
    plt.xticks(range(len(feature_names)), [feature_names[i] for i in indices], rotation=45, ha="right")
    plt.xlabel("Features", fontsize=12)
    plt.ylabel("Importance", fontsize=12)
    plt.tight_layout()
    plt.show()

# =============================================================================
# FIXED MODEL TRAINING WITH 80/20 SPLIT
# =============================================================================
def train_and_evaluate_model(dataset_path, modality_name, model_save_name):
    """Train model with 80/20 train-test split on ALL images"""
    print(f"\nüéØ Training {modality_name} model (80/20 split on all images)...")

    # Build dataset from ALL images
    dataset = []
    classes = ['normal', 'benign', 'malignant']

    for class_idx, class_name in enumerate(classes):
        class_path = os.path.join(dataset_path, class_name)
        if not os.path.exists(class_path):
            print(f"‚ö†Ô∏è Skipping {class_name} - folder not found")
            continue

        images = [f for f in os.listdir(class_path)
                 if f.lower().endswith(('.png', '.jpg', '.jpeg', '.pgm'))]

        print(f"üìÅ Processing {len(images)} {class_name} images...")

        for img_file in images:  # USE ALL IMAGES
            img_path = os.path.join(class_path, img_file)
            features = extract_enhanced_features(img_path, modality_name.lower())

            if features is not None:
                features['class'] = class_name
                dataset.append(features)

            # Progress indicator
            if len(dataset) % 100 == 0:
                print(f"  Processed {len(dataset)} images...")

    if len(dataset) < 30:
        print(f"‚ùå Insufficient data: {len(dataset)} samples")
        return None, None, None, None, None

    df = pd.DataFrame(dataset)
    print(f"\n‚úÖ Total dataset: {len(df)} samples")
    print(f"üìä Class distribution (BEFORE split):")
    print(df['class'].value_counts())

    # Prepare features
    feature_cols = [col for col in df.columns if col != 'class']
    X = df[feature_cols]
    y = df['class']

    # Encode labels
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    # ==================== FIXED 80/20 SPLIT ====================
    print(f"\nüîÑ Splitting data: 80% training, 20% testing...")

    # First split: 80% train, 20% test
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
    )

    print(f"üìä Original split:")
    print(f"  Training set: {X_train.shape[0]} samples ({X_train.shape[0]/len(df)*100:.1f}%)")
    print(f"  Testing set: {X_test.shape[0]} samples ({X_test.shape[0]/len(df)*100:.1f}%)")

    # Check class distribution in train and test
    print(f"\nüìä Class distribution in training set:")
    train_classes, train_counts = np.unique(y_train, return_counts=True)
    for cls_idx, count in zip(train_classes, train_counts):
        print(f"  {le.classes_[cls_idx]}: {count} samples")

    print(f"\nüìä Class distribution in testing set:")
    test_classes, test_counts = np.unique(y_test, return_counts=True)
    for cls_idx, count in zip(test_classes, test_counts):
        print(f"  {le.classes_[cls_idx]}: {count} samples")

    # Balance ONLY the training set with SMOTE
    print("\nüîÑ Balancing training set with SMOTE...")
    smote = SMOTE(random_state=42)
    X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

    print(f"üìä After SMOTE balancing:")
    print(f"  Balanced training set: {X_train_balanced.shape[0]} samples")
    balanced_classes, balanced_counts = np.unique(y_train_balanced, return_counts=True)
    for cls_idx, count in zip(balanced_classes, balanced_counts):
        print(f"  {le.classes_[cls_idx]}: {count} samples")

    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_balanced)
    X_test_scaled = scaler.transform(X_test)

    # Optimized Random Forest
    model = RandomForestClassifier(
        n_estimators=200,
        max_depth=20,
        min_samples_split=3,
        min_samples_leaf=1,
        max_features='sqrt',
        bootstrap=True,
        class_weight='balanced',
        random_state=42,
        n_jobs=-1
    )

    # Cross-validation on training set
    cv_scores = cross_val_score(model, X_train_scaled, y_train_balanced, cv=5)
    print(f"\nüìä 5-Fold Cross-Validation Accuracy: {cv_scores.mean():.3f} ¬± {cv_scores.std():.3f}")

    # Train model
    print("üèãÔ∏è Training model...")
    model.fit(X_train_scaled, y_train_balanced)

    # Predictions on test set (20% of original data)
    y_pred = model.predict(X_test_scaled)
    y_probs = model.predict_proba(X_test_scaled)

    # ==================== COMPREHENSIVE EVALUATION ====================
    print("\n" + "="*50)
    print(f"üìà COMPREHENSIVE EVALUATION - {modality_name}")
    print(f"   (Testing on 20% holdout dataset)")
    print("="*50)

    # 1. Basic Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    print(f"\nüìä Performance on TEST SET (20%):")
    print(f"   Accuracy:  {accuracy:.4f}")
    print(f"   Precision: {precision:.4f}")
    print(f"   Recall:    {recall:.4f}")
    print(f"   F1-Score:  {f1:.4f}")

    # 2. AUC Score
    try:
        auc_roc = roc_auc_score(y_test, y_probs, multi_class='ovr')
        print(f"   AUC-ROC:   {auc_roc:.4f}")
    except:
        auc_roc = 0.0
        print(f"   AUC-ROC:   N/A")

    # 3. Classification Report
    print(f"\nüìã Detailed Classification Report:")
    print(classification_report(y_test, y_pred, target_names=le.classes_, digits=4))

    # 4. Confusion Matrix
    print(f"\nüéØ Confusion Matrix (Test Set):")
    cm = confusion_matrix(y_test, y_pred)
    cm_df = pd.DataFrame(cm, index=le.classes_, columns=le.classes_)
    print(cm_df)

    # ==================== VISUALIZATIONS ====================
    print(f"\nüìä Generating visualizations...")

    # 1. Confusion Matrix Plot
    plot_confusion_matrix(y_test, y_pred, le.classes_,
                         f"Confusion Matrix - {modality_name} (Test Set)")

    # 2. ROC Curves
    plot_roc_curves(y_test, y_probs, le.classes_,
                   f"ROC Curves - {modality_name} (Test Set)")

    # 3. Feature Importance
    plot_feature_importance(model, feature_cols)

    # 4. Performance Summary
    print(f"\nüìä FINAL PERFORMANCE SUMMARY - {modality_name}")
    print("-"*50)
    print(f"Total dataset size: {len(df)} images")
    print(f"Training set (80%): {X_train_balanced.shape[0]} samples (after SMOTE)")
    print(f"Testing set (20%): {X_test.shape[0]} samples (original holdout)")
    print(f"Test Accuracy: {accuracy:.4f}")
    print(f"Test F1-Score: {f1:.4f}")
    print(f"Test AUC-ROC: {auc_roc:.4f}")
    print(f"CV Accuracy: {cv_scores.mean():.4f} ¬± {cv_scores.std():.4f}")

    # Save model
    joblib.dump(model, f'{model_save_name}.pkl')
    joblib.dump(scaler, f'{model_save_name}_scaler.pkl')
    joblib.dump(le, f'{model_save_name}_encoder.pkl')
    joblib.dump(feature_cols, f'{model_save_name}_features.pkl')

    print(f"\n‚úÖ Model saved as '{model_save_name}.pkl'")

    return model, scaler, le, feature_cols, {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'auc_roc': auc_roc,
        'cv_scores': cv_scores.tolist(),
        'total_samples': len(df),
        'train_samples': X_train_balanced.shape[0],
        'test_samples': X_test.shape[0],
        'test_distribution': dict(zip(le.classes_, test_counts))
    }

# =============================================================================
# PREDICTION FUNCTION
# =============================================================================
def predict_breast_cancer(image, modality):
    """Optimized prediction function"""
    try:
        if image is None:
            return "‚ùå Please upload an image", None

        # Select model
        if modality == "ultrasound":
            if 'model_us' not in globals() or model_us is None:
                return "‚ùå Ultrasound model not trained", None
            model, scaler, encoder, features = model_us, scaler_us, encoder_us, features_us
        else:
            if 'model_mg' not in globals() or model_mg is None:
                return "‚ùå Mammography model not trained", None
            model, scaler, encoder, features = model_mg, scaler_mg, encoder_mg, features_mg

        # Process image
        if isinstance(image, str):
            img_array = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
        else:
            img_array = np.array(Image.fromarray(image).convert('L'))

        # Save temporary file
        temp_path = os.path.join(tempfile.gettempdir(), "temp_img.png")
        cv2.imwrite(temp_path, img_array)

        # Extract features
        feature_dict = extract_enhanced_features(temp_path, modality)
        os.remove(temp_path)

        if feature_dict is None:
            return "‚ùå Feature extraction failed", None

        # Predict
        feature_vector = np.array([[feature_dict.get(col, 0) for col in features]])
        feature_vector_scaled = scaler.transform(feature_vector)

        prediction_encoded = model.predict(feature_vector_scaled)[0]
        probabilities = model.predict_proba(feature_vector_scaled)[0]
        prediction = encoder.inverse_transform([prediction_encoded])[0]
        confidence = probabilities[prediction_encoded]

        # Create result
        result_html = f"""
        <div style="font-family: Arial; padding: 20px;">
            <div style="background: {'#FF6B6B' if prediction=='malignant' else '#4ECDC4' if prediction=='benign' else '#45B7D1'};
                        padding: 20px; border-radius: 10px; color: white; text-align: center;">
                <h2>Prediction: {prediction.upper()}</h2>
                <p>Confidence: {confidence:.3f}</p>
            </div>

            <div style="margin-top: 20px;">
                <h3>Probability Distribution:</h3>
        """

        for i, cls in enumerate(encoder.classes_):
            prob = probabilities[i]
            result_html += f"""
                <div style="margin: 10px 0;">
                    <div style="display: flex; justify-content: space-between;">
                        <span>{cls.title()}</span>
                        <span>{prob:.3f}</span>
                    </div>
                    <div style="background: #e0e0e0; height: 20px; border-radius: 10px;">
                        <div style="background: {'#FF6B6B' if cls=='malignant' else '#4ECDC4' if cls=='benign' else '#45B7D1'};
                                    height: 100%; width: {prob*100}%; border-radius: 10px;"></div>
                    </div>
                </div>
            """

        result_html += "</div></div>"

        return result_html, confidence

    except Exception as e:
        return f"‚ùå Error: {str(e)}", None

# =============================================================================
# GRADIO INTERFACE
# =============================================================================
def create_interface():
    """Create clean Gradio interface"""
    with gr.Blocks(title="Breast Cancer Detection - 80/20 Split") as demo:

        gr.Markdown("# ü©∫ Breast Cancer Detection")
        gr.Markdown("### AI system trained on 80% of data, tested on 20%")

        with gr.Row():
            with gr.Column():
                modality = gr.Radio(
                    choices=["ultrasound", "mammography"],
                    label="Select Imaging Modality",
                    value="ultrasound"
                )
                image_input = gr.Image(
                    label="Upload Medical Image",
                    type="numpy",
                    image_mode="L"
                )
                analyze_btn = gr.Button("Analyze Image", variant="primary")

            with gr.Column():
                output_html = gr.HTML(label="Analysis Results")
                confidence = gr.Number(label="Confidence Score", visible=False)

        analyze_btn.click(
            predict_breast_cancer,
            inputs=[image_input, modality],
            outputs=[output_html, confidence]
        )

        gr.Markdown("---")
        gr.Markdown("""
        **Dataset Information (80/20 Split):**
        - Ultrasound: 1578 images (1262 train, 316 test)
        - Mammography: 1939 images (1551 train, 388 test)

        **Evaluation Metrics on 20% Test Set:**
        - Accuracy, Precision, Recall, F1-Score
        - AUC-ROC, Confusion Matrix
        - Feature Importance

        ‚ö†Ô∏è **Disclaimer:** For educational purposes only. Not for medical diagnosis.
        """)

    return demo

# =============================================================================
# MAIN EXECUTION
# =============================================================================
if __name__ == "__main__":
    print("üöÄ Starting Breast Cancer Detection System with 80/20 Split")
    print("="*60)

    # Initialize model variables
    model_us = scaler_us = encoder_us = features_us = eval_us = None
    model_mg = scaler_mg = encoder_mg = features_mg = eval_mg = None

    # Train ultrasound model
    if us_total >= 30:
        print("\n" + "="*60)
        print("TRAINING ULTRASOUND MODEL (80/20 Split)")
        print("="*60)
        model_us, scaler_us, encoder_us, features_us, eval_us = train_and_evaluate_model(
            "ultrasound_dataset", "Ultrasound", "ultrasound_model_80_20"
        )
    else:
        print("‚ö†Ô∏è Insufficient ultrasound data for training")

    # Train mammography model
    if mg_total >= 30:
        print("\n" + "="*60)
        print("TRAINING MAMMOGRAPHY MODEL (80/20 Split)")
        print("="*60)
        model_mg, scaler_mg, encoder_mg, features_mg, eval_mg = train_and_evaluate_model(
            "mammography_dataset", "Mammography", "mammography_model_80_20"
        )
    else:
        print("‚ö†Ô∏è Insufficient mammography data for training")

    # Display final summary
    print("\n" + "="*60)
    print("‚úÖ TRAINING COMPLETED (80/20 Split)")
    print("="*60)

    summary_data = []
    if eval_us is not None:
        summary_data.append({
            'Modality': 'Ultrasound',
            'Accuracy': f"{eval_us['accuracy']:.4f}",
            'F1-Score': f"{eval_us['f1_score']:.4f}",
            'AUC-ROC': f"{eval_us['auc_roc']:.4f}",
            'Train Samples': eval_us['train_samples'],
            'Test Samples': eval_us['test_samples']
        })

    if eval_mg is not None:
        summary_data.append({
            'Modality': 'Mammography',
            'Accuracy': f"{eval_mg['accuracy']:.4f}",
            'F1-Score': f"{eval_mg['f1_score']:.4f}",
            'AUC-ROC': f"{eval_mg['auc_roc']:.4f}",
            'Train Samples': eval_mg['train_samples'],
            'Test Samples': eval_mg['test_samples']
        })

    if summary_data:
        summary_df = pd.DataFrame(summary_data)
        print("\nüìä FINAL PERFORMANCE SUMMARY (80/20 Split):")
        print("-"*60)
        print(summary_df.to_string(index=False))

        # Calculate actual split percentages
        print("\nüìä ACTUAL DATA SPLIT:")
        print("-"*40)
        if eval_us:
            total_us = eval_us['total_samples']
            train_pct = (eval_us['train_samples'] / total_us) * 100
            test_pct = (eval_us['test_samples'] / total_us) * 100
            print(f"Ultrasound: {total_us} total ‚Üí {eval_us['train_samples']} train ({train_pct:.1f}%), {eval_us['test_samples']} test ({test_pct:.1f}%)")

        if eval_mg:
            total_mg = eval_mg['total_samples']
            train_pct = (eval_mg['train_samples'] / total_mg) * 100
            test_pct = (eval_mg['test_samples'] / total_mg) * 100
            print(f"Mammography: {total_mg} total ‚Üí {eval_mg['train_samples']} train ({train_pct:.1f}%), {eval_mg['test_samples']} test ({test_pct:.1f}%)")

    # Launch Gradio interface
    if model_us is not None or model_mg is not None:
        print("\nüöÄ Launching Gradio interface...")
        demo = create_interface()
        demo.launch(share=True, debug=False)
    else:
        print("\n‚ö†Ô∏è Cannot launch interface - no models trained successfully")
```
